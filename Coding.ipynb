```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats

global lineGraphX
global lineGraphY
lineGraphX = []
lineGraphY = []


def createLineGraph():

  global lineGraphX, lineGraphY

  print("Okay! Let's make a scatterplot/line graph.")
  print(
      "Input your points with x first and y second like this: (1, 2), (2, 4), (3, 6)..."
  )
  lineGraphInput = eval(input(""))
  lineGraphX, lineGraphY = zip(*lineGraphInput)
  lineGraphX = np.array(lineGraphX)
  lineGraphY = np.array(lineGraphY)
  print(
      "Do you want to make these data points \n 1. A line or \n 2. Scattered dots?"
  )
  lineOrScatter = input("")
  if lineOrScatter == "1":
    plt.plot(lineGraphX, lineGraphY)
  else:
    plt.plot(lineGraphX, lineGraphY, 'go')

  print("Would you like to create another line? (yes or no)")
  choice2_2_newLine = input("")
  choice2_2_newLine = choice2_2_newLine.lower()

  if choice2_2_newLine == "yes":
    print(
        "Input your points with x first and y second like this: (1, 2), (2, 4), (3, 6)..."
    )
    lineGraphInput_2 = eval(input(""))
    lineGraphX_2, lineGraphY_2 = zip(*lineGraphInput_2)
    plt.plot(lineGraphX_2, lineGraphY_2)

    print(
        "Do you want to make these data points \n 1. A line or \n 2. Scattered dots?"
    )
    lineOrScatter2 = input("")
    if lineOrScatter2 == "1":
      plt.plot(lineGraphX_2, lineGraphY_2)
    else:
      plt.plot(lineGraphX_2, lineGraphY_2, 'go')

    print("Would you like to create another line? (yes or no)")
    choice2_2_newLine_2 = input("")
    choice2_2_newLine_2 = choice2_2_newLine.lower()

    if choice2_2_newLine_2 == "yes":
      print(
          "Input your points with x first and y second like this: (1, 2), (2, 4), (3, 6)..."
      )
      lineGraphInput_3 = eval(input(""))
      lineGraphX_3, lineGraphY_3 = zip(*lineGraphInput_3)
      plt.plot(lineGraphX_3, lineGraphY_3)
      createLineGraph2(lineGraphX, lineGraphY)

      print(
          "Do you want to make these data points \n 1. A line or \n 2. Scattered dots?"
      )
      lineOrScatter3 = input("")
      if lineOrScatter3 == "1":
        plt.plot(lineGraphX_3, lineGraphY_3)
      else:
        plt.plot(lineGraphX_3, lineGraphY_3, 'go')

    else:
      createLineGraph2(lineGraphX, lineGraphY)

  else:
    createLineGraph2(lineGraphX, lineGraphY)


def createLineGraph2(lineGraphX, lineGraphY):

  print("Now let's add some basic parameters to the graph!")
  print("What do you want to title the graph?")
  lineGraphTitle = str(input(""))
  plt.title(lineGraphTitle)

  print("What do you want to label the x-axis as?")
  lineGraphXAxisTitle = str(input(""))
  plt.xlabel(lineGraphXAxisTitle)

  print("What do you want to label the y-axis as?")
  lineGraphYAxisTitle = str(input(""))
  plt.ylabel(lineGraphYAxisTitle)

  print(
      "Do you want to custome make the ticks? (Yes/no) If no, ticks will automatically be made."
  )
  lineGraphTicks = input("")
  lineGraphTicks = lineGraphTicks.lower()
  if lineGraphTicks == "yes":
    print(
        "What do you want the x-axis tick intervals to be set as? Ex. 0, 2, 4, 6"
    )
    lineGraphXAxisTicks = np.fromstring(input(""), dtype=float, sep=',')
    plt.xticks(lineGraphXAxisTicks)

    print(
        "What do you want the y-axis tick intervals to be set as? Ex. 0, 3, 6, 8, 15"
    )
    lineGraphYAxisTicks = np.fromstring(input(""), dtype=float, sep=',')
    plt.yticks(lineGraphYAxisTicks)
  else:
    pass

  print("Do you want to add the line of best fit? Yes or no?"
        )  #FIX REGRESSION LINE
  lineGraphRegressionLine = input("")
  lineGraphRegressionLine = lineGraphRegressionLine.lower()
  if lineGraphRegressionLine == "yes":

    m, b = np.polyfit(lineGraphX, lineGraphY, 1)
    plt.plot(lineGraphX, m * np.array(lineGraphX) + b)
  else:
    pass
  plt.show()


programLooper = True

while programLooper is True:

  print(
      "Hello, and welcome to the data management program. Created by Ethan. Would you like to: \n 1. Calculate measures of central tendency/spread? \n 2. Graph data? \n 3. Calculate if you have an outlier? or \n 4. Quit the program?"
  )
  choiceMain = input("")
  if choiceMain == "1":
    print(
        "Which measure of central tendency/spread would you like to calculate? \n 1. Mean? \n 2. Median? \n 3. Mode? \n 4. Range? \n 5. Standard Deviation? \n 6. Variance? \n 7. Quartiles? \n 8. Or Interquartile Range?"
    )
    choice1 = input("")
    if choice1 == "1":
      print("Enter your numbers here, and separate the numbers with a comma:")
      choice1_1 = np.fromstring(input(""), dtype=float, sep=',')
      choice1_1_ans = np.mean(choice1_1)
      print("The mean is: ", choice1_1_ans)
    elif choice1 == "2":  #Check what their definition of median is
      print("Enter your numbers here, and separate the numbers with a comma:")
      choice1_2 = np.fromstring(input(""), dtype=float, sep=',')
      choice1_2_ans = np.median(choice1_2)
      print("The median is: ", choice1_2_ans)
    elif choice1 == "3":
      print("Enter your numbers here, and separate the numbers with a comma:")
      choice1_3 = np.fromstring(input(""), dtype=float, sep=',')
      choice1_3_ans = scipy.stats.mode(choice1_3)
      print("The mode is: ", choice1_3_ans)
    elif choice1 == "4":
      print("Enter your numbers here, and separate the numbers with a comma:")
      choice1_4 = np.fromstring(input(""), dtype=float, sep=',')
      choice1_4_ans = max(choice1_4) - min(choice1_4)
      print("The range is:", choice1_4_ans)
    elif choice1 == "5":
      print("Enter your numbers here, and separate the numbers with a comma:")
      choice1_5 = np.fromstring(input(""), dtype=float, sep=',')
      choice1_5_ans = np.std(choice1_5)
      print("The standard deviation is: ", choice1_5_ans)
    elif choice1 == "6":
      print("Enter your numbers here, and seperate the numbers with a comma:")
      choice1_6 = np.fromstring(input(""), dtype=float, sep=',')
      choice1_6_ans = np.var(choice1_6)
      print("The variance is: ", choice1_6_ans)
    elif choice1 == "7":
      print("Enter your numbers here, and seperate the numbers with a comma:")
      choice1_7 = np.fromstring(input(""), dtype=float, sep=',')
      choice1_7_ans = np.quantile(choice1_7, [0.25, 0.5, 0.75])
      print("The quartiles are: ", choice1_7_ans)
    elif choice1 == "8":
      print("Enter your numbers here, and seperate the numbers with a comma:")
      choice1_8 = np.fromstring(input(""), dtype=float, sep=',')
      choice1_8_ans = np.quantile(choice1_8, [0.75]) - np.quantile(
          choice1_8, [0.25])
      print("The interquartile range is: ", choice1_8_ans)

  elif choiceMain == "2":
    print(
        "Okay, which graph would you like to make? \n 1. A box and whisker plot? \n 2. A line graph/scatterplot? or \n 3. A bar chart?"
    )
    choice2_whichGraph = input("")

    if choice2_whichGraph == "1":
      boxPlotLabel = ""  #FIX BOXPLOT
      print(
          "Alright! Enter your data set here, and separate it with a comma. Like this: 2, 19, 67, 35"
      )
      boxPlotData = np.fromstring(input(""), dtype=float, sep=',')

      print(
          "Do you want to make a horizontal or vertical boxPlot? (type 'horizontal' or 'vertical')"
      )
      boxPlotVertPrompt = input("")
      boxPlotVertPrompt = boxPlotVertPrompt.lower()
      if boxPlotVertPrompt == "horizontal":
        boxPlotVert = False
      else:
        boxPlotVert = True

      print(
          "Do you want to show the mean of the data? (Displayed as a blue triangle) (yes/no)"
      )
      boxPlotMeanPrompt = input("")
      boxPlotMeanPrompt = boxPlotMeanPrompt.lower()
      if boxPlotMeanPrompt == "yes":
        boxPlotMeanCheck = True
        meanMarker = dict(markerfacecolor='blue', marker='^')

      else:
        boxPlotMeanCheck = False
        meanMarker = dict(markerfacecolor='blue', marker='^')

      print("What do you want to title the graph?")
      boxPlotTitle = str(input(""))
      plt.title(boxPlotTitle)

      print("What do you want to label the x-axis as?")
      boxPlotXAxisTitle = str(input(""))
      plt.xlabel(boxPlotXAxisTitle)

      print("What do you want to label the y-axis as?")
      boxPlotYAxisTitle = str(input(""))
      plt.ylabel(boxPlotYAxisTitle)

      plt.boxplot(boxPlotData,
                  vert=boxPlotVert,
                  showmeans=boxPlotMeanCheck,
                  meanprops=meanMarker)
      plt.show()

    elif choice2_whichGraph == "2":  #MAKE IT ABLE TO HAVE DOTS/NOT HAVE DOTS
      createLineGraph()

    elif choice2_whichGraph == "3":  #ADD PARAMS LIKE TITLE
      barChartLabels = []
      barChartData = []

      print(
          "Input your bars with names first and heights second like this: A,2,Silo,4,Tophat,3... (notice how there are no spaces between the values.)"
      )
      barChartInput = input("")
      barChartInput2 = [
          float(num) if num.replace('.', '', 1).isdigit() else str(num)
          for num in barChartInput.split(',')
      ]
      for p in barChartInput2:
        if isinstance(p, str):
          barChartLabels.append(p)
        else:
          barChartData.append(p)

      barChartLabels = np.array(barChartLabels)
      barChartData = np.array(barChartData)
      plt.bar(barChartLabels, barChartData)

      print("Now let's add some basic parameters to the graph!")
      print("What do you want to title the graph?")
      barChartTitle = str(input(""))
      plt.title(barChartTitle)

      print("What do you want to label the x-axis as?")
      barChartXAxisTitle = str(input(""))
      plt.xlabel(barChartXAxisTitle)

      print("What do you want to label the y-axis as?")
      barChartYAxisTitle = str(input(""))
      plt.ylabel(barChartYAxisTitle)

      plt.show()

    else:
      "I'm sorry, but that is not a valid option. Please try again."

  elif choiceMain == "3":
    choice3_outlierCounter = 0
    print(
        "So you have chosen to calculate if there are any outliers in your data set. Input your data set below:"
    )
    choice3_Data = np.fromstring(input(""), dtype=float, sep=',')
    choice3_IQR = np.quantile(choice3_Data, [0.75]) - np.quantile(
        choice3_Data, [0.25])
    choice3_Q1 = np.quantile(choice3_Data, [0.25])
    choice3_Q3 = np.quantile(choice3_Data, [0.75])
    for n in choice3_Data:
      if n < choice3_Q1 - 1.5 * choice3_IQR or n > choice3_Q3 + 1.5 * choice3_IQR:
        print("YOU HAVE AN OUTLIER!: ", n)
        choice3_outlierCounter += 1
      else:
        print("Not an outlier:", n)
    print("Your data set had ", choice3_outlierCounter, " outliers.")

  elif choiceMain == "4":
    print("Exiting the program... Thank you for using my program! -Ethan")
    exit()
  else:
    print("I'm sorry, but that is not a valid option. Please try again.")
```
